## How GTJA Actually Extracts the 191 Factors

GuoTai JunAn alphas follow a repeatable **operator pipeline** on a wide OHLCV panel (date × ticker).

### 1. Universe & Panel
- Liquid A-share universe (CSI-style tickers).
- Wide matrices for `open`, `high`, `low`, `close`, `volume`, `amount`.

### 2. Derived Fields
- **VWAP** = amount / volume (fallback: typical price).
- **Return**, range_pct, body_pct, mid, pressure (close location in range).

### 3. Operator Vocabulary
| Axis | Operators |
|------|-----------|
| **Time-series (TS)** | DELTA, DELAY, SUM, MEAN, STD, MAX, MIN, CORR, TSRANK, SMA, WMA, DECAYLINEAR, REGBETA, REGRESI |
| **Cross-section (CS)** | RANK, SCALE |

TS runs per stock; CS runs across stocks on each date.

### 4. Factor Families (original constructions, same taxonomy)
1. **Momentum / reversal** — ranked deltas of price/volume fields.
2. **Volume–price coupling** — rolling corr of cross-sectionally ranked legs.
3. **Volatility** — range and return dispersion.
4. **VWAP deviation** — close vs VWAP, geometric mid vs VWAP.
5. **Conditional regime** — MA crossover gated by volume.
6. **Regression** — rolling beta/residual vs pseudo-market.
7. **Nonlinear rank mixes** — products/powers of CS ranks.

### 5. Extraction Pipeline
```
AkShare OHLCV → align wide panel → derive fields
    → 191 factor functions (operator DAGs) → date×ticker matrices
    → IC / quintile diagnostics (in-memory only)
```

> Formulas below are **independently constructed** — they mirror GTJA *methodology*, not copied formula sheets.

````python
import warnings
import time
from itertools import combinations, product
from typing import Callable

import akshare as ak
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
%matplotlib inline
plt.rcParams.update({"figure.dpi": 110, "font.size": 10})

N_FACTORS = 191
START, END = "2023-01-01", "2024-06-30"
SYMBOLS = ["600519", "000858", "601318", "600036", "000333", "601888", "300750"]
ADJUST = "qfq"
AK_SLEEP = 0.35
Panel = dict[str, pd.DataFrame]
FactorFn = Callable[[Panel], pd.DataFrame]
````

````python
# ── Panel utilities, operators, microstructure helpers ──

def safe_div(num, den):
    return (num / den.replace(0, np.nan)).replace([np.inf, -np.inf], np.nan)


def derive_fields(panel):
    out = dict(panel)
    if "vwap" not in out:
        if "amount" in out:
            out["vwap"] = safe_div(out["amount"], out["volume"])
        else:
            out["vwap"] = (out["high"] + out["low"] + out["close"]) / 3.0
    if "ret" not in out:
        out["ret"] = out["close"].pct_change()
    out["mid"] = (out["high"] + out["low"]) / 2.0
    out["range_pct"] = safe_div(out["high"] - out["low"], out["close"])
    out["body_pct"] = safe_div(out["close"] - out["open"], out["open"])
    return out


def align_panels(panels):
    idx = cols = None
    for df in panels.values():
        idx = df.index if idx is None else idx.union(df.index)
        cols = df.columns if cols is None else cols.union(df.columns)
    return {k: v.reindex(index=idx.sort_values(), columns=cols.sort_values()) for k, v in panels.items()}


def ts_delta(x, n=1):
    return x - x.shift(n)


def ts_delay(x, n=1):
    return x.shift(n)


def ts_sum(x, n):
    return x.rolling(n, min_periods=n).sum()


def ts_mean(x, n):
    return x.rolling(n, min_periods=n).mean()


def ts_std(x, n):
    return x.rolling(n, min_periods=n).std()


def ts_max(x, n):
    return x.rolling(n, min_periods=n).max()


def ts_min(x, n):
    return x.rolling(n, min_periods=n).min()


def ts_corr(a, b, n):
    return a.rolling(n, min_periods=n).corr(b)


def ts_log(x):
    return np.log(x.replace(0, np.nan))


def ts_rank(x, n):
    def _last_pct(w):
        s = pd.Series(w)
        return s.rank(pct=True).iloc[-1] if not s.isna().all() else np.nan
    return x.rolling(n, min_periods=n).apply(_last_pct, raw=False)


def ts_sma(x, n, m):
    arr = x.to_numpy(dtype=float)
    out = np.full_like(arr, np.nan)
    for j in range(arr.shape[1]):
        prev = np.nan
        for i in range(n - 1, len(arr[:, j])):
            xi = arr[i, j]
            if np.isnan(xi):
                continue
            prev = xi if np.isnan(prev) else (m * xi + (n - m) * prev) / n
            out[i, j] = prev
    return pd.DataFrame(out, index=x.index, columns=x.columns)


def ts_wma(x, n):
    w = np.array([0.9 ** i for i in range(n)][::-1], dtype=float)
    w /= w.sum()
    def _wma(window):
        ww = w[-len(window):] / w[-len(window):].sum()
        return np.nansum(window * ww)
    return x.rolling(n, min_periods=n).apply(_wma, raw=True)


def ts_decay_linear(x, n):
    w = np.arange(1, n + 1, dtype=float)
    w /= w.sum()
    def _decay(window):
        ww = w[-len(window):] / w[-len(window):].sum()
        return np.nansum(window * ww)
    return x.rolling(n, min_periods=n).apply(_decay, raw=True)


def ts_reg_beta(y, x, n):
    return safe_div(y.rolling(n, min_periods=n).cov(x), x.rolling(n, min_periods=n).var())


def ts_reg_resid(y, x, n):
    return y - ts_reg_beta(y, x, n) * x


def cs_rank(x):
    return x.rank(axis=1, pct=True, method="average")


def cs_scale(x):
    return x.div(x.abs().sum(axis=1).replace(0, np.nan), axis=0)


def pressure(close, high, low):
    return safe_div((close - low) - (high - close), high - low)


def shadow_ratio(high, low, close, open_):
    body_top = np.maximum(close, open_)
    body_bot = np.minimum(close, open_)
    rng = high - low
    upper = safe_div(high - body_top, rng)
    lower = safe_div(body_bot - low, rng)
    return upper, lower
````

````python
# ── AkShare loader (in-memory only, no cache) ──

def load_ashare_panel(symbols, start, end, adjust="qfq", sleep_sec=0.3):
    buckets = {k: [] for k in ("open", "high", "low", "close", "volume", "amount")}
    start_s, end_s = start.replace("-", ""), end.replace("-", "")

    for sym in symbols:
        raw = ak.stock_zh_a_hist(
            symbol=sym, period="daily",
            start_date=start_s, end_date=end_s, adjust=adjust,
        )
        if raw is None or raw.empty:
            print(f"  skip {sym}: empty")
            continue
        df = raw.rename(columns={
            "日期": "date", "开盘": "open", "收盘": "close",
            "最高": "high", "最低": "low", "成交量": "volume", "成交额": "amount",
        })
        df["date"] = pd.to_datetime(df["date"])
        df = df.set_index("date").sort_index()
        if "amount" not in df.columns:
            df["amount"] = df["close"] * df["volume"]
        for field in buckets:
            buckets[field].append(df[field].rename(sym))
        time.sleep(sleep_sec)

    if not buckets["close"]:
        raise RuntimeError("No data loaded — check symbols / network.")
    return align_panels({k: pd.concat(v, axis=1).sort_index() for k, v in buckets.items()})
````

````python
# ── Build 191 independently-constructed factors ──

def _f(p):
    return derive_fields(p)


def build_factor_registry():
    specs = []
    base = ["open", "high", "low", "close", "volume", "vwap", "ret"]

    # A: CS-RANK(TS-DELTA)
    for fld in base:
        for lag in (1, 2, 3, 5, 7, 10):
            def fn(p, fld=fld, lag=lag):
                f = _f(p)
                return cs_rank(ts_delta(f[fld], lag))
            specs.append(("momentum", f"cs_rank(delta {fld}, {lag}d)", fn))

    # B: -TS-CORR(CS-RANK legs)
    pairs = list(combinations(base, 2))
    for (fa, fb), win in product(pairs[:21], (5, 6, 8, 10, 12)):
        def fn(p, fa=fa, fb=fb, win=win):
            f = _f(p)
            return -ts_corr(cs_rank(ts_delta(f[fa], 1)), cs_rank(ts_delta(f[fb], 2)), win)
        specs.append(("vol_price_corr", f"-corr(rank d1 {fa}, d2 {fb}) {win}d", fn))

    # C: volatility
    for win in (5, 8, 10, 15, 20, 25, 30):
        def fn_r(p, win=win):
            f = _f(p)
            return cs_rank(safe_div(ts_std(f["range_pct"], win), ts_mean(f["range_pct"], win)))
        specs.append(("volatility", f"cs_rank(std/mean range {win}d)", fn_r))

        def fn_v(p, win=win):
            f = _f(p)
            return -cs_rank(ts_std(f["ret"], win))
        specs.append(("volatility", f"-cs_rank(ret std {win}d)", fn_v))

    # D: VWAP deviation
    for win in (3, 5, 10, 15, 20):
        def fn1(p, win=win):
            f = _f(p)
            dev = safe_div(f["close"] - f["vwap"], f["vwap"])
            return cs_rank(ts_mean(dev, win))
        specs.append(("vwap", f"cs_rank(mean close-vwap {win}d)", fn1))

        def fn2(p, win=win):
            f = _f(p)
            geo = np.sqrt(f["high"] * f["low"])
            return cs_rank(ts_delta(geo - f["vwap"], win))
        specs.append(("vwap", f"cs_rank(delta sqrt(HL)-vwap {win}d)", fn2))

    # E: pressure × volume
    for win in (4, 6, 8, 10, 12, 15):
        def fn(p, win=win):
            f = _f(p)
            pr = pressure(f["close"], f["high"], f["low"])
            return ts_sum(pr * f["volume"], win)
        specs.append(("flow", f"sum(pressure*vol {win}d)", fn))

    # F: conditional MA regime
    for ma_s, ma_l in ((5, 20), (8, 21), (10, 30), (3, 15)):
        def fn(p, ma_s=ma_s, ma_l=ma_l):
            f = _f(p)
            c, v = f["close"], f["volume"]
            short, long = ts_mean(c, ma_s), ts_mean(c, ma_l)
            vol_z = safe_div(v, ts_mean(v, 20))
            sig = pd.DataFrame(0.0, index=c.index, columns=c.columns)
            sig = sig.where(~(short > long), 1.0)
            sig = sig.where(~(short < long), -1.0)
            sig = sig.where(~((short <= long) & (vol_z >= 1.2)), 0.5)
            return sig
        specs.append(("regime", f"MA{ma_s}/{ma_l} vol-gated", fn))

    # G: regression
    for win in (10, 15, 20, 25, 30):
        def fn_b(p, win=win):
            f = _f(p)
            mkt = f["close"].mean(axis=1)
            mkt_df = pd.DataFrame({c: mkt for c in f["close"].columns}, index=f["close"].index)
            return cs_rank(ts_reg_beta(f["ret"], mkt_df.pct_change(), win))
        specs.append(("regression", f"cs_rank beta pseudo-mkt {win}d", fn_b))

        def fn_r(p, win=win):
            f = _f(p)
            return cs_rank(ts_reg_resid(f["close"], ts_delay(f["close"], 1), win))
        specs.append(("regression", f"cs_rank AR(1) resid {win}d", fn_r))

    # H: nonlinear rank mixes
    for win in (3, 5, 8, 10):
        def fn1(p, win=win):
            f = _f(p)
            return cs_rank(ts_delta(f["close"], win)) * cs_rank(ts_delta(f["volume"], win))
        specs.append(("nonlinear", f"rank(dclose)*rank(dvol) {win}d", fn1))

        def fn2(p, win=win):
            f = _f(p)
            r1 = cs_rank(f["body_pct"])
            r2 = cs_rank(ts_log(f["volume"]))
            return np.sign(r1 - 0.5) * (r2 ** 2)
        specs.append(("nonlinear", f"sign(body)*logvol rank sq", fn2))

    # I: decay / WMA / TSRANK composites
    for win in (5, 7, 10, 14, 20):
        def fn_d(p, win=win):
            f = _f(p)
            return cs_rank(ts_decay_linear(f["ret"], win))
        specs.append(("decay", f"cs_rank decay_linear ret {win}d", fn_d))

        def fn_w(p, win=win):
            f = _f(p)
            return cs_rank(ts_wma(f["close"], win) - f["close"])
        specs.append(("decay", f"cs_rank(wma-close) {win}d", fn_w))

        def fn_t(p, win=win):
            f = _f(p)
            return cs_rank(ts_rank(f["volume"], win))
        specs.append(("decay", f"cs_rank ts_rank vol {win}d", fn_t))

    # J: shadow / open-gap signals
    for win in (3, 5, 8, 10, 12):
        def fn_u(p, win=win):
            f = _f(p)
            u, _ = shadow_ratio(f["high"], f["low"], f["close"], f["open"])
            return cs_rank(ts_mean(u, win))
        specs.append(("microstructure", f"cs_rank upper shadow {win}d", fn_u))

        def fn_g(p, win=win):
            f = _f(p)
            gap = safe_div(f["open"] - ts_delay(f["close"], 1), ts_delay(f["close"], 1))
            return cs_rank(ts_sum(gap, win))
        specs.append(("microstructure", f"cs_rank sum gap {win}d", fn_g))

    # K: SMA recursive blends
    for n, m in ((7, 2), (10, 3), (12, 4), (15, 5)):
        def fn(p, n=n, m=m):
            f = _f(p)
            hl2 = (f["high"] + f["low"]) * 0.5
            prev = (ts_delay(f["high"], 1) + ts_delay(f["low"], 1)) * 0.5
            temp = hl2 - prev * safe_div(f["high"] - f["low"], f["volume"])
            return ts_sma(temp, n, m)
        specs.append(("blend", f"sma hl2 blend n={n} m={m}", fn))

    # Trim or pad to exactly 191
    if len(specs) < N_FACTORS:
        i = 0
        while len(specs) < N_FACTORS:
            fld = base[i % len(base)]
            lag = 1 + (i % 10)
            def fn_pad(p, fld=fld, lag=lag):
                f = _f(p)
                return cs_rank(ts_mean(ts_delta(f[fld], 1), lag))
            specs.append(("pad", f"cs_rank mean delta {fld} lag={lag}", fn_pad))
            i += 1
    specs = specs[:N_FACTORS]

    registry = {}
    for i, (fam, desc, fn) in enumerate(specs, start=1):
        registry[i] = {"id": i, "family": fam, "description": desc, "fn": fn}
    return registry


FACTOR_REGISTRY = build_factor_registry()
print(f"Built {len(FACTOR_REGISTRY)} factors")
fam_counts = pd.Series([v["family"] for v in FACTOR_REGISTRY.values()]).value_counts()
print(fam_counts.to_string())
````

````python
# ── Factor engine (GTJA-style extraction loop) ──

class AlphaEngine:
    def __init__(self, panel):
        self.panel = derive_fields(panel)

    def list_factors(self):
        return sorted(FACTOR_REGISTRY)

    def describe(self, factor_id):
        meta = FACTOR_REGISTRY[factor_id]
        return {"id": factor_id, "family": meta["family"], "description": meta["description"]}

    def compute(self, factor_ids=None):
        if factor_ids is None:
            factor_ids = self.list_factors()
        return {fid: FACTOR_REGISTRY[fid]["fn"](self.panel) for fid in factor_ids}

    def compute_all(self):
        return self.compute()


def daily_ic(factor, fwd_ret, method="spearman"):
    ics = []
    for dt in factor.index.intersection(fwd_ret.index):
        f, r = factor.loc[dt], fwd_ret.loc[dt]
        ok = f.notna() & r.notna()
        if ok.sum() < 3:
            continue
        ics.append((dt, f[ok].corr(r[ok], method=method)))
    return pd.Series(dict(ics)).sort_index() if ics else pd.Series(dtype=float)


def forward_return(close, horizon=5):
    return close.shift(-horizon) / close - 1
````

````python
# ── Step 1: Load A-share data (AkShare, in-memory) ──
print(f"Downloading {len(SYMBOLS)} A-shares via AkShare …")
raw_panel = load_ashare_panel(SYMBOLS, START, END, adjust=ADJUST, sleep_sec=AK_SLEEP)
print("Panel shapes:", {k: v.shape for k, v in raw_panel.items()})
raw_panel["close"].tail()
````

````python
# ── Step 2: Derive fields & compute all 191 factors ──
engine = AlphaEngine(raw_panel)
t0 = time.time()
factors = engine.compute_all()
elapsed = time.time() - t0
print(f"Computed {len(factors)} factors in {elapsed:.1f}s")
for fid in [1, 2, 50, 100, 191]:
    meta = engine.describe(fid)
    last = factors[fid].iloc[-1].dropna()
    print(f"  alpha_{fid:03d} [{meta['family']}]: {meta['description']}")
    print(f"    last values: {last.to_dict()}")
````

````python
# ── Step 3: IC diagnostics (in-memory, no files saved) ──
close = engine.panel["close"]
fwd = forward_return(close, horizon=5)

rows = []
for fid in sorted(factors):
    ic = daily_ic(factors[fid], fwd)
    rows.append({
        "factor": f"alpha_{fid:03d}",
        "family": FACTOR_REGISTRY[fid]["family"],
        "ic_mean": ic.mean(),
        "ic_std": ic.std(),
        "ir": ic.mean() / ic.std() if ic.std() else np.nan,
        "hit_rate": (ic > 0).mean() if len(ic) else np.nan,
    })

ic_df = pd.DataFrame(rows).sort_values("ir", ascending=False, na_position="last")
print("Top 10 factors by IR:")
display(ic_df.head(10))
````

````python
# ── Step 4: Visualizations (inline only) ──
show_ids = [1, 2, 50, 100, 191]
fig, axes = plt.subplots(2, 3, figsize=(14, 7))
axes = axes.ravel()

# Price panel
ax = axes[0]
for sym in close.columns:
    ax.plot(close.index, close[sym], label=sym, alpha=0.8)
ax.set_title("A-share close prices (qfq)")
ax.legend(fontsize=7, ncol=2)
ax.tick_params(axis="x", rotation=30)

# Factor heatmap (last 30 days, showcase factors)
ax = axes[1]
heat = pd.concat({f"α{fid:03d}": factors[fid].iloc[-30:].mean(axis=1) for fid in show_ids}, axis=1)
im = ax.imshow(heat.T, aspect="auto", cmap="RdBu_r")
ax.set_yticks(range(len(show_ids)))
ax.set_yticklabels([f"α{i:03d}" for i in show_ids])
ax.set_title("Factor cross-sectional mean (last 30d)")
plt.colorbar(im, ax=ax, fraction=0.046)

# IC by family
ax = axes[2]
ic_df.boxplot(column="ic_mean", by="family", ax=ax, rot=45)
ax.set_title("IC distribution by family")
fig.suptitle("")

# Cumulative IC for top factor
ax = axes[3]
top_fid = int(ic_df.iloc[0]["factor"].split("_")[1])
ic_top = daily_ic(factors[top_fid], fwd)
ax.plot(ic_top.index, ic_top.cumsum())
ax.set_title(f"Cumulative IC — alpha_{top_fid:03d}")
ax.tick_params(axis="x", rotation=30)

# Scatter: factor vs forward return (last date)
ax = axes[4]
dt = factors[1].dropna(how="all").index[-1]
ax.scatter(factors[1].loc[dt], fwd.loc[dt])
for sym in close.columns:
    ax.annotate(sym, (factors[1].loc[dt, sym], fwd.loc[dt, sym]), fontsize=7)
ax.set_xlabel("alpha_001")
ax.set_ylabel("5d forward return")
ax.set_title(f"Scatter @ {dt.date()}")

# IR bar chart (top 15)
ax = axes[5]
top15 = ic_df.dropna(subset=["ir"]).head(15)
ax.barh(top15["factor"], top15["ir"])
ax.invert_yaxis()
ax.set_title("Top 15 factors by IR")
ax.set_xlabel("IR")

plt.tight_layout()
plt.show()
print("Done — all results kept in memory, nothing saved to disk.")
````

## Summary

This notebook reproduces the **GTJA extraction process**:

1. **Load** A-share OHLCV via AkShare into wide panels.
2. **Derive** VWAP, returns, range/body ratios, pressure.
3. **Apply** nested TS + CS operators to build 191 alpha signals.
4. **Evaluate** with daily IC / IR — all in memory.

Each of the 191 factors belongs to a conceptual family (momentum, vol-price corr, volatility, etc.) but uses **original operator compositions** rather than verbatim published GTJA formulas.
